```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
#Let's plt the log function
x = np.arange(-5, 5, 0.1)
h = lambda z : np.log(z)
plt.plot(x, h(x), alpha=0.5)
plt.show()
```

```python
#Let's plot the negative log function
#Let's plt the log function
x = np.arange(-5, 5, 0.1)
h = lambda z : -np.log(z)
plt.plot(x, h(x), alpha=0.5)
plt.show()
```
