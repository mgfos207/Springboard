```python
df[['beds', 'bedrooms']] = df[['beds', 'bedrooms']].fillna(1)
df.drop(columns=['reviews_per_month', 'review_scores_value', 'review_scores_location', 'review_scores_communication', 'review_scores_checkin', 'review_scores_cleanliness', 'review_scores_accuracy', 'review_scores_rating', 'latitude', 'longitude', 'minimum_minimum_nights','maximum_minimum_nights','minimum_maximum_nights','maximum_maximum_nights','minimum_nights_avg_ntm','maximum_nights_avg_ntm'], inplace=True)
```

```python
df.head()
```

Reporoduce the last steps from the preprocessing notebook (test train split and scaling values)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.loc[:, ~df.columns.isin(['percentage_aval_30'])]
y = df['percentage_aval_30']
```

```python
y.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=43)
```

Standardize (scale) some of the feature output

```python
scale_features = ['number_of_reviews','maximum_nights','minimum_nights','price','number_of_reviews','number_of_reviews_ltm','number_of_reviews_l30d']
for feature in scale_features:
    scaler = StandardScaler()
    #train the scaler based on training data
    df_feat = np.asarray(X_train[feature])
    df_feat = df_feat.reshape(-1, 1)
    X_train[feature] = scaler.fit_transform(df_feat)
    #apply fit data model to test data
    df_feat_test = np.asarray(X_test[feature])
    df_feat_test = df_feat_test.reshape(-1, 1)
    X_test[feature] = scaler.transform(df_feat_test)
```
